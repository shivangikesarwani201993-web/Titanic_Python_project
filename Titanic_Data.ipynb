1.5) Handling missing Values

a) Filling median value of age column in the missing values of the same column.

```python
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
```

b) Filling mode value of Embarked column in place of missing value of same column.

```python
embarked_mode = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(embarked_mode)
```

c) Creating a new column called "Has_cabin" where 0 indicates no cabin and 1 indicates has cabin.

```python
df['Has_Cabin'] = df['Cabin'].notna().astype(int)
```

```python
df.drop('Cabin', axis =1 , inplace = True)
```

Bivariate Analysis

a.) Does gender affect the chances of survival? (Survived vs Sex)

```python
df.groupby('Sex')['Survived'].value_counts()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'Sex', y = 'Survived', data = df).set_title("Survival by Sex")
plt.show()
```

b.) How does age influence survival rate? (Survived vs Agegroups)

```python
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
(pd.crosstab(df['AgeGroup'], df['Survived'],  normalize = 'index')*100)
```

```python
(pd.crosstab(df['AgeGroup'], df['Survived'],  normalize = 'index')*100).plot(kind = 'bar', title = "Age group influence on survival count", figsize =(12,8) )
plt.show()
```

j) Survival Rate by Title?

```python
df['Name']
```

```python
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand = False)
```

```python
df['Title']
```

```python
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

```

```python
df['Title'] = df['Title'].replace('Mlle', 'Miss')
df['Title'] = df['Title'].replace('Ms', 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')
```
